```python
# Loading the Dependences
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
# Loading the dataset from scikit-learn liobrary
load_data = sklearn.datasets.load_boston()
print(type(load_data))
print(load_data)
```

```python
# Loading the datasets to a Pandas DataFrame
boston_data = pd.DataFrame(load_data.data,columns=load_data['feature_names'])
boston_data.head()
```

```python
# Adding target(prices) column to boston_data
boston_data['prices'] = load_data.target
boston_data.head()
```

```python
# Shape of the datasets
boston_data.shape
```

```python
# Statistical information of the datasets
boston_data.describe()
```

```python
# Printing data types of each columns of the datasets
boston_data.info()
```

```python
# Finding missing values
boston_data.isnull().sum()
```

```python
# Finding Duplicate values
boston_data.duplicated().sum()
```

```python
# Understanding the relation among each columns
correlation = boston_data.corr()

# Visualizing correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,annot=True,fmt=".2f",cmap="viridis")
```

```python
# Separating the dataset into Dependent and Independent variables
X = boston_data.drop(columns=['prices'],axis=1)
Y = boston_data['prices']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)

print(X.shape,x_train.shape,x_test.shape)
```

```python
# Fitting the Training datasets to Support Vector Machine model
from xgboost import XGBRegressor
model = XGBRegressor()
model.fit(x_train,y_train)
```

# Evaluation


```python
# training data accuracy
model.score(x_train,y_train)
```

```python
# test data accuracy
model.score(x_test,y_test)
```

```python
# Evaluate the training datasets model
x_train_prediction = model.predict(x_train)
score_1 = r2_score(y_train,x_train_prediction)
print("R Squared Error on Train Datasets:",score_1)

score_2 = mean_absolute_error(y_train,x_train_prediction)
print("Mean Absuluter Error on Train Datasets:",score_2)
```

```python
# Evaluate the test datasets model
x_test_prediction = model.predict(x_test)
score_3 = r2_score(y_test,x_test_prediction)
print("R Squared Error on Test Datasets:",score_3)

score_4 = mean_absolute_error(y_test,x_test_prediction)
print("Mean Absuluter Error on Test Datasets:",score_2)
```

# Visualizing the Actual prices vs Predicted prices

```python
plt.scatter(y_train,x_train_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual prices vs Predicted prices")
plt.show()
```
